```python
# importing sklearn for model selection and preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Splitting the data into X and Y

X = df.drop('TenYearCHD', axis=1)
y = df['TenYearCHD']
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Rescaling The Dataset

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
print(scaled_X_train)
print(scaled_X_test)
```

# 1. Logistic Regression

```python
# Importing the Model

from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
log_reg.fit(scaled_X_train,y_train)
```

```python
# Importing the Necessary Metrics For Model Evaluation

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

ypre=log_reg.predict(scaled_X_test)
```

```python
print('accracy score is:',accuracy_score(y_test,ypre))
```

```python
# Error Rate

error_rate=1-accuracy_score(y_test,ypre)
print("error rate is:",error_rate)
```

```python
# Confusion Matrix

print("Confusion Matrix \n",confusion_matrix(y_test,ypre))
```

```python
# Classification Report

print("Classification report \n ",classification_report(y_test,ypre))
```

# 2. K- Nearest Neighbors

```python
# Importing The Model

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7) # here n_neighbors is k 
knn.fit(scaled_X_train,y_train)
```

```python
y_knn_pre = knn.predict(scaled_X_test)
```

```python
# Importing Necessary Metrics

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
# Accuracy Score

print('accracy score is:',accuracy_score(y_knn_pre,y_test))
```

```python
# Error Rate

error_knn = 1 - accuracy_score(y_knn_pre,y_test)
print("error rate is:",error_knn)
```

```python
# Confusion Matrix

print("Confusion matrix \n",confusion_matrix(y_test,y_knn_pre))
```

```python
# Classification Report

print("Classification Report \n",classification_report(y_test,y_knn_pre))
```

# 3. Decision Tree

```python
# Model Imputation

from sklearn.tree import DecisionTreeClassifier
```

```python
# Decision Tree With Criterion As Gini

dt_gini = DecisionTreeClassifier(criterion = "gini",
            random_state = 10,max_depth=3, min_samples_leaf=5)
dt_gini.fit(scaled_X_train,y_train)
```

```python
y_gini_pred = dt_gini.predict(scaled_X_test)
```

```python
# Accuracy Score

print('accracy score is:',accuracy_score(y_gini_pred,y_test))
```

```python
# Confusion Matrix

print("Confusion matrix \n",confusion_matrix(y_test,y_gini_pred))
```

```python
# Classification Report

print("Classification Report \n",classification_report(y_test,y_gini_pred))
```

```python
# Decision Tree With Criterion As Entropy

dt_entropy = DecisionTreeClassifier(
            criterion = "entropy", random_state = 10,
            max_depth = 3, min_samples_leaf = 5)
dt_entropy.fit(scaled_X_train,y_train)
```

```python
y_entropy_pred = dt_entropy.predict(scaled_X_test)
```

```python
# Accuracy Score

print('accracy score is:',accuracy_score(y_entropy_pred,y_test))
```

```python
# Confusion Matrix

print("Confusion Matrix \n",confusion_matrix(y_test,y_entropy_pred))
```

```python
# Classification Report

print("classification report \n",classification_report(y_test,y_entropy_pred))
```
