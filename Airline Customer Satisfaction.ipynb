```python
X = df.drop('satisfaction',axis=1)
```

```python
y = df['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

# 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis_reg = LogisticRegression()
logis_reg.fit(scaled_X_train,y_train)
```

```python
ypred=logis_reg.predict(scaled_X_test)
ypred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print('accracy score is:',accuracy_score(y_test,ypred))
```

```python
error_rate=1-accuracy_score(y_test,ypred)
print("error rate is:",error_rate)
```

```python
print("Confusion Matrix \n",confusion_matrix(y_test,ypred))
```

```python
print("Classification report \n ",classification_report(y_test,ypred))
```

# 3. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_gini = DecisionTreeClassifier(criterion = "gini",
            random_state = 10,max_depth=5, min_samples_leaf=5)
dt_gini.fit(scaled_X_train,y_train)
```

```python
y_gini_pred = dt_gini.predict(scaled_X_test)
```

```python
print('accracy score is:',accuracy_score(y_gini_pred,y_test))
```

```python
print("Classification Report \n",classification_report(y_test,y_gini_pred))
```

```python
dt_entropy = DecisionTreeClassifier(
            criterion = "entropy", random_state = 10,
            max_depth = 5, min_samples_leaf = 5)
dt_entropy.fit(scaled_X_train,y_train)
```

```python
y_entropy_pred = dt_entropy.predict(scaled_X_test)
```

```python
print('accracy score is:',accuracy_score(y_entropy_pred,y_test))
```

```python
print("Confusion Matrix \n",confusion_matrix(y_test,y_entropy_pred))
```

```python
print("classification report \n",classification_report(y_test,y_entropy_pred))
```
